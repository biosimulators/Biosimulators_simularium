```python
import numpy as np
import matplotlib.pyplot as plt

# Example signal (replace this with your actual signal)
t = np.linspace(0, 1, 400, endpoint=False)
signal = np.sin(2 * np.pi * 50 * t) + 0.5 * np.sin(2 * np.pi * 80 * t)

# Perform FFT
fft_output = np.fft.fft(signal)

# Get magnitudes
magnitudes = np.abs(fft_output)

# Compression: Zero out small magnitudes
threshold = 0.10 * np.max(magnitudes)  # for example, 10% of the max value
fft_output[magnitudes < threshold] = 0

# Perform Inverse FFT
reconstructed_signal = np.fft.ifft(fft_output)

# Plotting
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.title("Original Signal")
plt.plot(t, signal)

plt.subplot(2, 1, 2)
plt.title("Reconstructed Signal After Compression")
plt.plot(t, reconstructed_signal.real)  # Only plot the real part

plt.tight_layout()
plt.show()
```

```python
reconstructed_signal.real[-1], signal[-1]
```

```python
compressed_non_zero_count = np.count_nonzero(reconstructed_signal)
sig = np.count_nonzero(signal)

print(f"Original non-zero frequency components: {sig}")
print(f"Compressed non-zero frequency components: {compressed_non_zero_count}")
```

```python
import trimesh

points = [[0, 0, 0], [0, 0, 1], [0, 1, 0]]
faces = [[0, 1, 2]]
tmesh = trimesh.Trimesh(points, faces=faces, process=False)
mesh = pv.wrap(tmesh)
print(mesh)
```

```python
import vtk

points = vtk.vtkPoints()
p = [1.0, 2.0, 3.0]
vertices = vtk.vtkCellArray()
pid = points.InsertNextPoint(p)
vertices.InsertNextCell(1)
vertices.InsertCellPoint(pid)
point = vtk.vtkPolyData()
point.SetPoints(points)
point.SetVerts(vertices)
mesh = pv.wrap(point)
print(mesh)
```

```python
mesh.plot()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data (replace with your simulation outputs)
# Assuming the first column is time and the second column is the variable of interest
data = np.array([[23, 42, 24, 535], [23, 35435, 234, 234]])  # Your simulation output array of shape (2, 4)

# Fourier Transform
time = data[:, 0]
signal = data[:, 1]
fft_result = np.fft.fft(signal)
frequencies = np.fft.fftfreq(len(time), time[1] - time[0])

# Plot the spectrum
plt.figure(figsize=(10, 4))
plt.plot(frequencies, np.abs(fft_result))
plt.title("Frequency Spectrum")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```
